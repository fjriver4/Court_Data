```python
import pandas as pd

# Load both CSV files into DataFrames
df_filed = pd.read_csv('JP22_EvictionFiled_Jan20Sep1524_RAW.csv')
#df_judge = pd.read_csv('Post_DEBTPilot_Batch_Judge1_Sept10_RAW.csv')

# Combine the two DataFrames, prioritizing df_judge in case of duplicate 'Case Number'
#df_combined = pd.concat([df_filed, df_judge], ignore_index=True)

# Remove duplicates, keeping the last occurrence (which will be from df_judge if there's a conflict)
#df_combined = df_combined.drop_duplicates(subset='Case Number', keep='last')
df_combined = df_filed
# Display the first few rows of the deduplicated DataFrame
df_combined.info()
```

```python
import pandas as pd

# Load the CSV file into a DataFrame
df = df_combined

# Specify the column indices to be removed
columns_to_remove = [1, 2, 4, 5] + list(range(9, 14)) + list(range(15, 20)) + list(range(21, 26)) + list(range(27, 59)) + list(range(66, 71))

# Remove the columns by index
df_cleaned = df.drop(df.columns[columns_to_remove], axis=1)

# Display the first few rows of the cleaned DataFrame
df_cleaned.info()
```

```python
import pandas as pd

# Remove any trailing spaces from column names
df_cleaned.columns = df_cleaned.columns.str.strip()

# Create new binary columns based on the specified conditions

# JFP and JFD logic
df_cleaned['JFP'] = df_cleaned.apply(lambda row: 1 if row['Plaintiff Name'] == row['Judgment In Favor Of'] else (0 if row['Defendant Name'] == row['Judgment In Favor Of'] else None), axis=1)
df_cleaned['JFD'] = df_cleaned.apply(lambda row: 1 if row['Defendant Name'] == row['Judgment In Favor Of'] else (0 if row['Plaintiff Name'] == row['Judgment In Favor Of'] else None), axis=1)

# PATT logic
df_cleaned['PATT'] = df_cleaned['Plaintiff Atty Name'].apply(lambda x: 1 if pd.notna(x) and x.strip() != '' else 0)

# DATT logic
df_cleaned['DATT'] = df_cleaned['Defendant Atty Name'].apply(lambda x: 1 if pd.notna(x) and x.strip() != '' else 0)

# Display the first few rows to verify the results
df_cleaned.info()
```

```python
df_cleaned.to_csv('JP22_EV_FILED_2020_92024_CLEAN.csv', index=False)
```

```python
import pandas as pd

# Load the CSV file into a new DataFrame
df_sept10 = pd.read_csv('Sept10OdysseyDebtClaimColumns.csv')

# Specify the columns to remove by their index
columns_to_remove = [0, 3, 4, 5, 6, 10]

# Remove the columns by index
df_sept10_cleaned = df_sept10.drop(df_sept10.columns[columns_to_remove], axis=1)

# Remove trailing spaces from column names
df_sept10_cleaned.columns = df_sept10_cleaned.columns.str.strip()

# Remove the time from 'Event Date/Time', keeping only the date (MM/DD/YYYY)
df_sept10_cleaned['Event Date/Time'] = df_sept10_cleaned['Event Date/Time'].apply(lambda x: x.split()[0] if isinstance(x, str) else x)

# Display the first few rows of the cleaned DataFrame
df_sept10_cleaned.head()
```

```python
import pandas as pd
import numpy as np

# Create new columns based on 'Event Type' conditions and set corresponding 'Event Date/Time' with NaN for missing data

# Answer Due Date
df_sept10_cleaned['Answer Due Date'] = df_sept10_cleaned.apply(
    lambda row: row['Event Date/Time'] if row['Event Type'] == 'Case - Answer Due' else np.nan, axis=1
)

# Answer Filed Date
df_sept10_cleaned['Answer Filed Date'] = df_sept10_cleaned.apply(
    lambda row: row['Event Date/Time'] if row['Event Type'] == 'Case - Answer Filed' else np.nan, axis=1
)

# Service Return Date
df_sept10_cleaned['Service Return Date'] = df_sept10_cleaned.apply(
    lambda row: row['Event Date/Time'] if row['Event Type'] == 'Case - Service Return' else np.nan, axis=1
)

# Alt Service Authorized Date
df_sept10_cleaned['Alt Service Authorized Date'] = df_sept10_cleaned.apply(
    lambda row: row['Event Date/Time'] if row['Event Type'] == 'Case - Order-Authorizing Alternative Service 501.2' else np.nan, axis=1
)

# Display the first few rows of the updated DataFrame to verify the new columns
df_sept10_cleaned.head()
```

```python
# Remove the specified columns
df_sept10_cleaned = df_sept10_cleaned.drop(columns=['Event Type', 'Event Status', 'Event Date/Time', 'Event Status Date'])

# Display the first few rows of the updated DataFrame to verify the columns were removed
df_sept10_cleaned.head()

```

```python
import pandas as pd

# Group by 'Case #' and merge rows, giving priority to non-NaN values
df_sept10_cleaned = df_sept10_cleaned.groupby('Case #', as_index=False).first()

# Display the first few rows of the merged DataFrame to verify the changes
df_sept10_cleaned.info()
```

```python
import pandas as pd

# Check the column names to identify any potential discrepancies
print(df_cleaned.columns)
print(df_sept10_cleaned.columns)

# If the 'Case Number' column has a different name, rename it accordingly in both dataframes
df_cleaned = df_cleaned.rename(columns={'Case #': 'Case Number'})
df_sept10_cleaned = df_sept10_cleaned.rename(columns={'Case #': 'Case Number'})

# Ensure both 'Case Number' columns are of the same type (string/object)
df_cleaned['Case Number'] = df_cleaned['Case Number'].astype(str)
df_sept10_cleaned['Case Number'] = df_sept10_cleaned['Case Number'].astype(str)

# Select only the necessary columns from df_sept10_cleaned
columns_to_merge = ['Case Number', 'Answer Due Date', 'Answer Filed Date', 'Service Return Date', 'Alt Service Authorized Date']

# Merge df_cleaned with df_sept10_cleaned on 'Case Number'
df_cleaned = pd.merge(df_cleaned, df_sept10_cleaned[columns_to_merge], on='Case Number', how='left')

# If '_x' and '_y' columns exist, combine them, otherwise keep the newly merged columns
if 'Answer Due Date_x' in df_cleaned.columns and 'Answer Due Date_y' in df_cleaned.columns:
    df_cleaned['Answer Due Date'] = df_cleaned['Answer Due Date_y'].combine_first(df_cleaned['Answer Due Date_x'])
    df_cleaned['Answer Filed Date'] = df_cleaned['Answer Filed Date_y'].combine_first(df_cleaned['Answer Filed Date_x'])
    df_cleaned['Service Return Date'] = df_cleaned['Service Return Date_y'].combine_first(df_cleaned['Service Return Date_x'])
    df_cleaned['Alt Service Authorized Date'] = df_cleaned['Alt Service Authorized Date_y'].combine_first(df_cleaned['Alt Service Authorized Date_x'])

    # Drop the extra _x and _y columns
    df_cleaned = df_cleaned.drop(columns=['Answer Due Date_x', 'Answer Due Date_y', 'Answer Filed Date_x', 'Answer Filed Date_y', 
                                          'Service Return Date_x', 'Service Return Date_y', 'Alt Service Authorized Date_x', 'Alt Service Authorized Date_y'])

# Display the first few rows of the updated df_cleaned to verify the changes
df_cleaned.info()
```

```python
# Convert all NaN values to None
df_cleaned = df_cleaned.where(pd.notna(df_cleaned), None)

# Display the first few rows to verify the change
df_cleaned.head()
```

```python
import pandas as pd

# Ensure the relevant date columns are in datetime format
df_cleaned['Case File Date'] = pd.to_datetime(df_cleaned['Case File Date'], errors='coerce')
df_cleaned['Answer Filed Date'] = pd.to_datetime(df_cleaned['Answer Filed Date'], errors='coerce')
df_cleaned['Answer Due Date'] = pd.to_datetime(df_cleaned['Answer Due Date'], errors='coerce')
df_cleaned['Disposition Date'] = pd.to_datetime(df_cleaned['Disposition Date'], errors='coerce')

# Create 'TimeToAnswer' column: number of days between 'Case File Date' and 'Answer Filed Date'
df_cleaned['TimeToAnswer'] = (df_cleaned['Answer Filed Date'] - df_cleaned['Case File Date']).dt.days

# Create 'AnswerLate' column: 1 if 'Answer Filed Date' is after 'Answer Due Date', 0 if before, NaN if either is blank
df_cleaned['AnswerLate'] = df_cleaned.apply(
    lambda row: 1 if pd.notna(row['Answer Filed Date']) and pd.notna(row['Answer Due Date']) and row['Answer Filed Date'] > row['Answer Due Date']
    else (0 if pd.notna(row['Answer Filed Date']) and pd.notna(row['Answer Due Date']) and row['Answer Filed Date'] <= row['Answer Due Date'] else None),
    axis=1
)

# Create 'TimeToDispose' column: number of days between 'Case File Date' and 'Disposition Date'
df_cleaned['TimeToDispose'] = (df_cleaned['Disposition Date'] - df_cleaned['Case File Date']).dt.days

# Display the first few rows to verify the new columns
df_cleaned.head()
```

```python
# Export df_cleaned as a CSV file
df_cleaned.to_csv('Pre_DebtPilot_Cases_Sept10_CLEAN.csv', index=False)
```
