```python
import requests
import pandas as pd
from io import StringIO
from datetime import datetime
from dateutil.relativedelta import relativedelta
from concurrent.futures import ThreadPoolExecutor, as_completed
from tqdm import tqdm

# Configuration
court_ids = [305, 310, 315, 320, 325, 330, 335, 340, 345, 350, 355, 360, 365, 370, 375, 380]
case_types = ['ADMIN','BRSB','BOND','DD','CTA','DEBT','DCD','DLS','EV','FJ','HGL','ODL','OEPR','RR','SC','SSP','TAX',
             'TOW','TURN','WRITG','WR','WRU','SEQ','WRJ'] # Debt Claim
start_date = "2022-01-01"
end_date = "2025-12-31"
format = "tab"
base_url = "https://jpwebsite.harriscountytx.gov/PublicExtracts/GetExtractData"
extract_ids = ['6', '7']  # NEW: run all three extracts

def generate_date_ranges(start_date, end_date):
    start = datetime.strptime(start_date, "%Y-%m-%d")
    end = datetime.strptime(end_date, "%Y-%m-%d")
    date_ranges = []
    current = start
    while current <= end:
        start_of_month = current.replace(day=1)
        end_of_month = (start_of_month + relativedelta(months=1)) - relativedelta(days=1)
        date_ranges.append((start_of_month.strftime("%m/%d/%Y"), end_of_month.strftime("%m/%d/%Y")))
        current += relativedelta(months=1)
    return date_ranges

def download_data(court, case_type, extract_id, fdate, tdate):
    try:
        params = {
            'extractCaseType': 'CV',
            'extract': extract_id,   # NEW
            'court': court,
            'casetype': case_type,
            'format': format,
            'fdate': fdate,
            'tdate': tdate
        }
        headers = {'Accept': 'text/tab-separated-values'}
        response = requests.get(base_url, params=params, headers=headers, timeout=60)
        response.raise_for_status()

        if "attachment" in response.headers.get("Content-Disposition", ""):
            data = pd.read_csv(StringIO(response.text), sep="\t", dtype=str)

            if not data.empty and 'Case Number' in data.columns:
                # Drop any repeated header rows that sometimes show up in the body
                data = data[data['Case Number'] != 'Case Number']

                # Keep provenance if you want to audit later
                data['case_type'] = case_type
                data['extract_id'] = extract_id
                return data

        return None

    except Exception as e:
        print(f"[ERROR] Court {court} | Case Type {case_type} | Extract {extract_id} | {fdate}{tdate}")
        print("Reason:", e)
        return None

# Create all tasks (court x case_type x extract_id x month)
date_ranges = generate_date_ranges(start_date, end_date)
tasks = [
    (court, case_type, extract_id, start, end)
    for court in court_ids
    for case_type in case_types
    for extract_id in extract_ids
    for (start, end) in date_ranges
]

# Download in parallel with progress bar
all_data = []
with ThreadPoolExecutor(max_workers=8) as executor:
    futures = {executor.submit(download_data, *task): task for task in tasks}
    for future in tqdm(as_completed(futures), total=len(futures), desc="Downloading"):
        result = future.result()
        if result is not None and not result.empty:
            all_data.append(result)

# Combine and de-duplicate by Case Number across ALL extracts/months
if all_data:
    combined_df = pd.concat(all_data, ignore_index=True)

    # Drop rows with the same Case Number value (keep first occurrence)
    #combined_df = combined_df.drop_duplicates(subset=['Case Number'], keep='first')

    combined_df.to_csv('Harris_CivilHearingsFilings_Jan22Dec25_RAW.csv', index=False)
    print(combined_df.info())
else:
    print("No data was successfully downloaded.")
```
